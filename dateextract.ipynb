```python
import re  #import regular expression

# Open the file that you want to search and extract the dates from
try:
    f = open("doc.txt", "r")

# Will contain the entire content of the file as a string
    str = f.read()
    str = str.upper()
    
    print(' -------Date expression in text--------- ')

    #extract date dd/mm/YYYY format
    all = re.findall(r"[\d]{1,2}/[\d]{1,2}/[\d]{4}", str)
    for s in all:
        print(s)
        
    #extract dd-mm-yyyy date format
    all = re.findall(r"[\d]{1,2}-[\d]{1,2}-[\d]{2,4}", str)
    for s in all:
        print(s)
        
    #extract dd-MON-yyyy or dd-MON-YY date format
    all = re.findall(r"([\d]{1,2}\s(?:JAN|FEB|MAR|APR|JUN|JUL|AUG|SEPT|OCT|DEC|NOV)\s[\d]{2,4})", str)
    for s in all:
        print(s)
        
    #extract dd-MONTH-yyyy
    all = re.findall(r"([\d]{1,2}\s(JANUARY|FEBRUARY|MARCH|APRIL|MAY|JUNE|JULY|AUGUST|SEPTEMBER|OCTOBER|NOVEMBER|DECEMBER)\s[\d]{4})", str)
    for s in all:
        print(s)
        
    #extract date like dd MONTH
    all = re.findall(r"((?:JANUARY|FEBRUARY|MARCH|APRIL|MAY|JUNE|JULY|AUGUST|SEPTEMBER|OCTOBER|NOVEMBER|DECEMBER)\s[\d]{2}[\s]{1})", str)  
    for s in all:
        print(s)
        
    #extract year in BC,AD and BS upto 6 digits
    all = re.findall(r"([\d]{1,6}\s(?:BC|AD|BS))", str)  
    for s in all:
        print(s)
        
    print('\n -------Time expression in text--------- ')
     
    #extract time in AM and PM
    all = re.findall(r"([\d]{1,2}:[\d]{1,2}\s(?:AM|PM))", str)  
    for s in all:
        print(s)
        
     #extract time in 24 HR
    all = re.findall(r"([\d]{1,2}:[\d]{1,2}\s(?:[^AM ^PM][\w]{0,7}))",str) 
    print('\n')
    for s in all:
            print(s)
        
      #extract time hour and minute
    all = re.findall(r"(\s[\d]{1,2}\s(?:MINUTES|HOURS|HOUR|MINUTE|SECONDS|SECOND))", str)
    print('\n')
    for s in all:
        print(s)
            
    if all:
        print("\n ****This is all match possible****\n")
    else:
        print("\n ****No match found****\n")

except:
    print("The file is not present at its location. Please chk it once")
```
